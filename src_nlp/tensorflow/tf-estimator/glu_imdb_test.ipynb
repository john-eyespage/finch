$$ Y = X \otimes [1 - \sigma( \space Conv1D_2(X) \space )] \space + \space Conv1D_1(X) \otimes \sigma( \space Conv1D_2(X) \space ) $$

$$ Y = X + ( \space Conv1D_1(X) - X \space ) \otimes \sigma( \space Conv1D_2(X) \space ) $$

$$ \text{Assume} \space \space Conv1D_1(X) \space \space \text{learns} \space \space ( \space Conv1D_1(X) - X \space) $$

$$ Y = X + Conv1D_1(X) \otimes \sigma( \space Conv1D_2(X) \space ) $$

Related Papers:
* [Language Modeling with Gated Convolutional Networks](https://arxiv.org/abs/1612.08083)
* [Convolutional Sequence to Sequence Learning](https://arxiv.org/abs/1705.03122)

```python
import tensorflow as tf
import numpy as np
```

```python
VOCAB_SIZE = 5000
MAX_LEN = 400
BATCH_SIZE = 32
K_SIZE = 3
EMBED_DIM = 50
N_CLASS = 2
N_EPOCH = 2
LR = {'start': 5e-3, 'end': 5e-4, 'steps': 1500}
```

```python
def gated_linear_unit(x, d_rate):
    c = tf.layers.conv1d(inputs = x,
                         filters = 2*EMBED_DIM,
                         kernel_size = K_SIZE,
                         dilation_rate = d_rate,
                         padding = 'same')
    c1, c2 = tf.split(c, 2, -1)
    x += c1 * tf.sigmoid(c2)
    return x


def forward(x, mode):
    is_training = (mode == tf.estimator.ModeKeys.TRAIN)
    
    x = tf.contrib.layers.embed_sequence(x, VOCAB_SIZE, EMBED_DIM)
    x = tf.layers.dropout(x, 0.2, training=is_training)
    
    for d_rate in [1, 2, 4]:
        x = gated_linear_unit(x, d_rate)
    
    x = tf.reduce_max(x, 1)
    x = tf.layers.flatten(x)
    
    x = tf.layers.dropout(x, 0.2, training=is_training)
    x = tf.layers.dense(x, EMBED_DIM, tf.nn.relu)
    logits = tf.layers.dense(x, N_CLASS)
    return logits


def model_fn(features, labels, mode):
    logits = forward(features, mode)
    
    if mode == tf.estimator.ModeKeys.PREDICT:
        preds = tf.argmax(logits, -1)
        return tf.estimator.EstimatorSpec(mode, predictions=preds)
    
    if mode == tf.estimator.ModeKeys.TRAIN:
        global_step = tf.train.get_global_step()

        lr_op = tf.train.exponential_decay(
            LR['start'], global_step, LR['steps'], LR['end']/LR['start'])

        loss_op = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(
            logits=logits, labels=labels))

        train_op = tf.train.AdamOptimizer(lr_op).minimize(
            loss_op, global_step=global_step)

        lth = tf.train.LoggingTensorHook({'lr': lr_op}, every_n_iter=100)
        
        return tf.estimator.EstimatorSpec(
            mode=mode, loss=loss_op, train_op=train_op, training_hooks=[lth])
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=VOCAB_SIZE)
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, MAX_LEN)
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test, MAX_LEN)

estimator = tf.estimator.Estimator(model_fn)

for _ in range(N_EPOCH):
    estimator.train(tf.estimator.inputs.numpy_input_fn(
        x = X_train, y = y_train,
        batch_size = BATCH_SIZE,
        shuffle = True))
    y_pred = np.fromiter(estimator.predict(tf.estimator.inputs.numpy_input_fn(
        x = X_test,
        batch_size = BATCH_SIZE,
        shuffle = False)), np.int32)
    print("\nValidation Accuracy: %.4f\n" % (y_pred==y_test).mean())
```
